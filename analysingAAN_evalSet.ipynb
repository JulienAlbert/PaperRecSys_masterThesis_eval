```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
aan_dataset_path = "./evaloff3/data/aan_full.json"
```

```python
with open(aan_dataset_path) as f:
    aan_dataset = json.load(f)
    
len(aan_dataset)
```

```python
fields = set()

for reference in aan_dataset:
    for key in reference.keys():
        fields.add(key)
        
fields
```

```python
fields_count = {}
for field in fields:
    fields_count[field] = 0
    
for reference in aan_dataset:
    for field in  fields:
        if reference[field]:
            fields_count[field] += 1
            
fields_count
```

```python
for ref in aan_dataset:
    ref['title_word_count'] = len(ref['title'].split())
    ref['abstract_word_count'] = len(ref['abstract'].split())
    ref['author_count'] = len(ref['author'])
    ref['year'] = int(ref['year'])
    ref['citation_count'] = len(ref['citations'])
    ref['reference_count'] = len(ref['references'])
    ref['link_count'] = len(ref['citations']) + len(ref['references'])
    ref['fos_count'] = len(ref['fos'])
    
aan_df = pd.DataFrame(aan_dataset)
```
